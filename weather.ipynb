Machine Learning Model

```python
core_weather["target"]=core_weather.shift(-1)["temp_max"]
#We are trying to predict tomorrow's temperature
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[:-1,:].copy()
#remove the last row
```

```python
core_weather
```

```python
from sklearn.linear_model import Ridge
#Ridge regression reduces overfitting
reg = Ridge(alpha=.1)
```

```python
predictors = ["precipitation","snow","snow_depth", "temp_max", "temp_min"]
```

```python
# train = core_weather.sample(frac=0.8)
# test = core_weather.drop(train.index)

# print(f"No. of training examples: {train.shape[0]}")
# print(f"No. of testing examples: {test.shape[0]}")
```

```python
train = core_weather.loc[:"2008-12-31"]
test = core_weather.loc["2009-01-01":]
print(f"No. of training examples: {train.shape[0]}")
print(f"No. of testing examples: {test.shape[0]}")
# The above does a 80-20 split
```

```python
train
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
```
